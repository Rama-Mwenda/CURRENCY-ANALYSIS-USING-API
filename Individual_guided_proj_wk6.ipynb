## *ANALYSING FOREX RATES WITH THE FRANKFURTER API* ##

### Frankfurter is an open-source API for current and historical foreign exchange rates published by the European Central Bank.

```python
#Import the relevant libraries
import pandas as pd
import requests
from pandas import json_normalize
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


print('libraries loaded successdfy')
```

```python
#saving API link to url variable
URL = 'https://api.frankfurter.app/2013-01-01..2022-12-31'
```

```python
#GET data
response = requests.get(URL)
#read the .json file response
data = response.json()
#convert it to a dataframe
df = pd.DataFrame(data)
df
```

```python
#normalize rates column using json_normalize
df_rates = json_normalize(df['rates'])
df_rates
```

```python
# reset the index for the df_rates
df = df.reset_index()
#drop rates
df = df.drop('rates', axis = 1)
#df dataframe
df = pd.concat([df, df_rates], axis = 1)

df
```

```python
#converting columns to datetime
df['index'] = pd.to_datetime(df['index'])
df['start_date'] = pd.to_datetime(df['start_date'])
df['end_date'] = pd.to_datetime(df['end_date'])

#confirming the datatypes 
df.info()
```

**The dataset has 38 columns and 522 entries.**
**There are 4 different datatypes in the dataset, the float and int carry the currency prices and the object is the name of the base currency, while datetime carries the dates**

**QUESTION ONE: HOW HAS THE EXCHANGE RATE OF THE EURO AGAINST THE $ DOLLAR CHANGED OVER TIME**

```python
#define the plot size
plt.figure(figsize = (15, 5))
#make the lineplot
sns.lineplot(x='index', y='USD', data=df,)
#remove the borders
sns.despine(bottom = False, left = False)
#define the max and min points in the trend
max_point = df.loc[df['USD'].idxmax()]
min_point = df.loc[df['USD'].idxmin()]
# Annotate max point
plt.annotate(f'Max: {max_point["USD"]:.2f}', 
             xy=(max_point['index'], max_point['USD']),
             xytext=(-40, 10),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->", lw=1))
# Annotate min point
plt.annotate(f'Min: {min_point["USD"]:.2f}', 
             xy=(min_point['index'], min_point['USD']),
             xytext=(20, 20),
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->", lw=1))
#add chart title
plt.title(f'EUR-USD EXCHANGE TREND OVER TIME\n 2013 - 2023')
#show the plot
plt.show()
```

**QUESTION THREE: BETWEEN THE GREAT BRITISH POUND AND THE JAPANESE YEN DETERMINE WHICH COUNTRY'S CURRENCY IS MORE VOLATILE**


*To calculate currency volatility, you need to:*
 - Choose a time period and record the highest and lowest exchange rates for the currency pair you are interested in.
 - Calculate the difference between the highest and lowest exchange rates for each day or interval within the time period.
 - Add all of the differences together and divide this number by the total number of differences you recorded. This will give you the average difference or the mean deviation.
 - Divide the standard deviation by the average difference and multiply the result by 100. This will give you the volatility rating as a percentage. The higher the percentage, the more volatile the currency pair is.

```python
#the highest and lowest exchange rates for gbp and yen
gbp_high = df['GBP'].max()
gbp_low = df['GBP'].min()
jpy_high = df['JPY'].max()
jpy_low = df['JPY'].min()
```

```python
#difference between highest and lowest for each day gbp
daily_high_gbp = gbp_high - df['GBP']
daily_low_gbp = df['GBP'] - gbp_low
#sum of all daily differences
diff_gbp = daily_high_gbp.sum() + daily_low_gbp.sum()
```

```python
#difference between highest and lowest for each day jpy
daily_high_jpy = jpy_high - df['JPY']
daily_low_jpy = df['JPY'] - jpy_low
#sum of all daily differences
diff_jpy = daily_high_jpy.sum() + daily_low_jpy.sum()
```

```python
#add all the differences and divide by the count for the mean deviation
gbp_mean_dev = diff_gbp / df['GBP'].count()
jpy_mean_dev = diff_jpy / df['JPY'].count()
```

```python
#converting to percentage
gbp_vol = (df['GBP'].std() / gbp_mean_dev)*100
jpy_vol = (df['JPY'].std() / jpy_mean_dev)*100
```

```python
#saving volatilities to a dictionary
vol_comp = {'GBP':gbp_vol, 'JPY':jpy_vol}
plot = pd.DataFrame(vol_comp.items())
plot
```

```python
# Create an array with the colors to use
colors = ["#492E87", "#37B5B6"]
#plotting the volatility
ax = sns.barplot(x=0, y =1, data = plot, hue = 0, palette = colors)
# Add data labels to the bars 
for c in ax.containers:     
    ax.bar_label(c,) 
#add title
plt.xlabel("CURRENCY")
plt.ylabel("% VOLATILITY")
plt.title(f"PERCENTAGE VOLATILITY OF GBP AND JPY AGAINST THE EUR\n 2013 - 2023")
plt.show()
```

The chart shows that GBP has higher volatility than JPY by approximately 2.6 percentage points.
